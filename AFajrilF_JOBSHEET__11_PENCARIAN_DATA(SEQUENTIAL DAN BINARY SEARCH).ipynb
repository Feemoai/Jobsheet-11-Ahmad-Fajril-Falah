Praktikum 01 – Sequential Search pada Data Angka


---



```python
def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return f"Data {target} ditemukan di indeks {i}"
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(sequential_search(angka, 30))
print(sequential_search(angka, 100))
```

Praktikum 02 – Binary Search pada Data Angka

---



```python
def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return f"Data {target} ditemukan di indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"Data {target} tidak ditemukan"

angka = [10, 20, 30, 40, 50]
print("Data:", angka)
print(binary_search(angka, 40))
print(binary_search(angka, 100))
```

Praktikum 05 – Perbandingan Waktu Sequential vs Binary Search

```python
import time

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

data = list(range(1, 100001))
target = 99999

start = time.time()
sequential_search(data, target)
end = time.time()
print("Sequential Search waktu:", end - start, "detik")

start = time.time()
binary_search(data, target)
end = time.time()
print("Binary Search waktu:", end - start, "detik")
```

Praktikum 06 – Kombinasi Sorting dan Searching

---



```python
import random

def insertion_sort(data):
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1
        while j >= 0 and data[j] > key:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key
    return data

def binary_search(data, target):
    left, right = 0, len(data)-1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return f"{target} ditemukan di indeks {mid}"
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return f"{target} tidak ditemukan"

data = [random.randint(1, 100) for _ in range(15)]
print("Data acak sebelum diurutkan:", data)
urut = insertion_sort(data)
print("Data setelah diurutkan:", urut)
target = urut[7]  # ambil angka ke-8 sebagai target
print(binary_search(urut, target))
```

4.	Variasi Binary Search

---



```python
print("Ahmad Fajril Falah")
print("4. Variasi Binary Search\n")
import random

def binary_search_counter(data, target):
    left = 0
    right = len(data) - 1
    jumlah_langkah = 0  #variabel penghitungnya

    while left <= right:
        jumlah_langkah += 1  # Tambah 1 ditiap kali looping

        mid = (left + right) // 2

        if data[mid] == target:
            return f"Angka {target} DITEMUKAN di indeks {mid}. (Butuh {jumlah_langkah} langkah)"

        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return f"Angka {target} TIDAK DITEMUKAN. (Total {jumlah_langkah} langkah)"

data = [random.randint(1, 25) for _ in range(15)]
data.sort() #agar datanya urut
target = random.choice(data) # saya mau tagetnya juga acak

print(f"Data (Terurut):{data}")
print(f"Target yang dicari:{target}\n")
hasil = binary_search_counter(data, target)
print(hasil)
```

6.	Studi Kasus Barang Toko

---



```python
print("Ahmad Fajril Falah")
print("6.Studi Kasus Barang Toko\n")
def cari_batas_atas(data, budget):
    left = 0
    right = len(data) - 1
    index_ditemukan = -1  # jika tdk ada yg cocok

    while left <= right:
        mid = (left + right) // 2
        harga_barang = data[mid][1]

        if harga_barang <= budget:
            # Jika harga masuk budget, simpan index ini
            index_ditemukan = mid
            # coba cari di sebelah KANAN (mahal)
            # apakah ada yg masuk budget
            left = mid + 1
        else:
            # Jika harga terlalu mahal, cari di sebelah KIRI (barang lebih murah)
            right = mid - 1

    return index_ditemukan

# Filtering Logic

barang_toko = []
print("=== FILTER BARANG SESUAI BUDGET ===")

try:
    # masukan data
    jumlah = int(input("Mau input berapa barang? "))
    for i in range(jumlah):
        print(f"Barang ke-{i+1}:")
        nama = input("  Nama : ")
        harga = int(input("  Harga: "))
        barang_toko.append((nama, harga))

    # sorting data
    barang_toko.sort(key=lambda x: x[1])

    print("\n" + "="*30)
    print("DAFTAR SEMUA BARANG:")
    for item in barang_toko:
        print(f"- {item[0]}: Rp{item[1]}")

    print("="*30)
    # input budget maksimal
    budget = int(input("\nMasukkan budget maksimal kamu: "))
    # batas index dengan binary search
    batas_index = cari_batas_atas(barang_toko, budget)
    # Hasilnya
    print("-" * 30)

    if batas_index != -1:
        print(f"Daftar barang dengan harga di bawah Rp{budget}:")
        # Loop dari index 0 sampai batas_index
        for i in range(batas_index + 1):
            item = barang_toko[i]
            print(f"{i+1}. {item[0]} - Rp{item[1]}")
    else:
        print(f"Tidak ada barang yang harganya di bawah Rp{budget}.")


except ValueError:
    print("Error: Masukkan angka yang benar.")
```

7.	Analisis Kompleksitas

---



```python
print("Ahmad Fajril Falah")
print("7.Analisis Kompleksitas\n")
import time
import random

def sequential_search(data, target):
    for i in range(len(data)):
        if data[i] == target:
            return i
    return -1

def binary_search(data, target):
    left, right = 0, len(data) - 1
    while left <= right:
        mid = (left + right) // 2
        if data[mid] == target:
            return mid
        elif data[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# logic pengukurannya
print(f"{'UKURAN DATA':<15} | {'SEQUENTIAL (ns)':<20} | {'BINARY SEARCH (ns)':<20}")
print("-" * 65)

ukuran_data = [100, 1000, 10000]

for n in ukuran_data:
    # data dummy urut dari 0 sampai n-1
    data = list(range(n))

    # tentukan target yg tdk ada, misal n + 100
    target = n + 100

    # UKUR SEQUENTIAL
    start_seq = time.perf_counter_ns()  # Mulai stopwatch (nanosecond)
    sequential_search(data, target)
    end_seq = time.perf_counter_ns()    # Stop stopwatch
    waktu_seq = end_seq - start_seq

    # UKUR BINARY
    # data sudah urut, jadi aman untuk Binary Search
    start_bin = time.perf_counter_ns()
    binary_search(data, target)
    end_bin = time.perf_counter_ns()
    waktu_bin = end_bin - start_bin

    # HASILNYA disni
    print(f"{n:<15} | {waktu_seq:<20} | {waktu_bin:<20}")

print("-" * 65)
print("Catatan: ns = nanosecond (1 detik = 1.000.000.000 ns)")
```

8.	Proyek Mini Pencarian Nilai Mahasiswa

```python
def binary_search_simple(data, target_nilai):
    left = 0
    right = len(data) - 1

    while left <= right:
        mid = (left + right) // 2
        nilai_tengah = data[mid][1] #ambil nilai di tengah

        if nilai_tengah == target_nilai:
            return mid
        elif nilai_tengah < target_nilai:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# --- PROGRAM UTAMA ---
mahasiswa = []
print("=== Ahmad Fajril Falah")
print("=== PROGRAM PENCARIAN NILAI ===")

# INPUT JUMLAH DATA
jumlah = int(input("Mau input berapa mahasiswa? "))

# INPUT DATA MAHASISWA
print(f"Silakan masukkan data untuk {jumlah} orang:")
for i in range(jumlah):
    nama = input(f"Nama Mahasiswa {i+1}: ")
    nilai = int(input(f"Nilai {nama}: "))
    mahasiswa.append((nama, nilai))

# logic SORTING
mahasiswa.sort(key=lambda x: x[1])

print("\nData setelah diurutkan berdasarkan nilai:")
print(mahasiswa)

# logic SEARCHING
cari = int(input("\nMasukkan nilai yang ingin dicari: "))
hasil = binary_search_simple(mahasiswa, cari)

# hasilnya
if hasil != -1:
    print(f"KETEMU! Mahasiswa bernama {mahasiswa[hasil][0]} memiliki nilai {cari}.")
else:
    print(f"Tidak ada mahasiswa dengan nilai {cari}.")
```
